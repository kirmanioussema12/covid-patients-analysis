# <span style="color:Green">P1: Load the dataset.</span>

```python
import csv,sys
import numpy as np
file_path=r"C:\Users\MSI\Desktop\assignment 2\e760480e-1f95-4634-a923-98161cfb02fa.csv"
```

```python
#loading the Dataset using the csv module which is a built-in module in Python for handling CSV files.
# We open the file in read mode with UTF-8 encoding, create a DictReader to read the CSV data into a list of dictionaries,
# and then print the first  row of the dataset to verify that it has been loaded correctly.
data=[]
with open(file_path, 'r',encoding="utf-8-sig") as file :
    reader = csv.DictReader(file)
    for row in reader:
        data.append(row)
print(data[0])
        
```

# <span style="color:Green">P2: How many unique values exist in the 'oh_region' column? What are the unique values in the 'oh_region' column? How many records exist per 'oh_region'?</span>

```python
#extracting the 'oh_region' column from the dataset and converting it into a NumPy array. 
oh_region=np.array([row['oh_region'] for row in data])
# We then use the np.unique function to find the unique values in the 'oh_region' column and count their occurrences.
unique_values,count=np.unique(oh_region,return_counts=True)
# Finally, we print the number of unique values, the unique values themselves, and the count of records for each region.
print("The number of unique values in the regions is",len(unique_values))
print("\nthe unique values in the regions are",unique_values)
print("\nRecords per region:")
for i in range(len(unique_values)):
    print(unique_values[i],":",count[i])
```

# <span style="color:Green">P3: What is the total number of hospitalizations? What is the average number of hospitalizations per day?</span>

```python
#extracting the 'hospitalizations' column from the dataset and converting it into a NumPy array.
hospitilizations=np.array([row['hospitalizations'] for row in data])
# We then convert the 'hospitalizations' values from strings to floats using np.array and astype(float).
hospitilizations=hospitilizations.astype(int)
#we calculate the total number of hospitalizations by summing the values in the 'hospitalizations' array using np.sum and print the result.
sum_hospitilizations=np.sum(hospitilizations)
print("\nThe total number of hospitalizations is :",sum_hospitilizations)

# Extract dates from the dataset and convert them into a NumPy array. We then group the hospitalizations by date and sum them across regions for each date.
dates = np.array([row["date"] for row in data], dtype=str)
# Group by date and sum hospitalizations across regions for each date
unique_dates, inverse_idx = np.unique(dates, return_inverse=True)
daily_totals = np.bincount(inverse_idx, weights=hospitilizations)
print("\nDaily total hospitalizations:", daily_totals)
# We then calculate the total number of hospitalizations by summing the values in the 'daily_totals' array using np.sum and print the result.
avg_hospitalizations_per_day = daily_totals.mean()
print("\nAverage hospitalizations per day:", avg_hospitalizations_per_day)


```

# <span style="color:Green">P4: What are the top 5 days with the highest number of hospitalizations?</span>

```python
#Codes of P4 here
# Get indices of top 5 hospitalization values
top5_idx = np.argsort(hospitilizations)[-5:][::-1]

print("Top 5 days with highest hospitalizations:\n")

for i in top5_idx:
    print("Date:", dates[i], "| Hospitalizations:", hospitilizations[i])
```

# <span style="color:Green">P5: Which month(s) had the highest number of hospitalizations?</span>

```python
# Extract month and year from the date strings and create a new array that contains only the month and year information
months = np.array([d[5:7] for d in dates])   # keeps "YYYY-MM"

# Group by month
unique_months, inverse_idx = np.unique(months, return_inverse=True)
monthly_totals = np.bincount(inverse_idx, weights=hospitilizations)
# Find max
max_value = np.max(monthly_totals)
max_months = unique_months[monthly_totals == max_value]

print("Month(s) with highest hospitalizations (across all years):")
for m in max_months:
    print("Month:", m)

print("\nHighest total:", max_value)
```

 # <span style="color:Green">P6: Are there any seasonal (Winter, Spring, Summer, Fall) (Average)patterns in COVID-19 hospitalizations?</span>

```python
# Define seasons based on month
import numpy as np
from datetime import datetime
# Define seasons
seasons = {12: 'Winter', 1: 'Winter', 2: 'Winter',
           3: 'Spring', 4: 'Spring', 5: 'Spring',
           6: 'Summer', 7: 'Summer', 8: 'Summer',
           9: 'Fall', 10: 'Fall', 11: 'Fall'}

# Create storage
season_values = {'Winter': [], 'Spring': [], 'Summer': [], 'Fall': []}

# Assign hospitalizations to seasons
for date_str, hosp in zip(dates, hospitilizations):
    try:
        month = int(date_str[5:7])  # extract month as integer
        season = seasons[month]
        season_values[season].append(hosp)
    except:
        continue

# Compute seasonal averages
season_averages = {}
for season in ['Winter', 'Spring', 'Summer', 'Fall']:
    if season_values[season]:
        avg = np.mean(season_values[season])
        season_averages[season] = avg
        print(f"{season}: {avg:.2f}")
    else:
        print(f"{season}: No Data")

# Highest average season
if season_averages:
    highest_season = max(season_averages, key=season_averages.get)
    print(f"\nSeason with Highest Average Hospitalizations: {highest_season} ({season_averages[highest_season]:.2f})")
```

# <span style="color:Green">P8: What is the average number of ICU patients on ventilators ('icu_current_covid_vented') per region?</span>

```python
#Codes of P8 here
import numpy as np

region_key = "oh_region"
vent_key = "icu_current_covid_vented"

# Extract regions
regions = np.array([row[region_key] for row in data], dtype=str)

# Clean ventilator values (replace '.' with NaN)
vent = np.array([
    float(row[vent_key]) if row[vent_key] not in ('.', '', 'NA') else np.nan
    for row in data
], dtype=float)

# Group by region
unique_regions, inverse_idx = np.unique(regions, return_inverse=True)

# Sum ventilated ICU per region
total_vent = np.bincount(inverse_idx, weights=np.nan_to_num(vent))

# Count valid (non-NaN) entries per region
valid_counts = np.bincount(inverse_idx, weights=~np.isnan(vent))

# Compute average
avg_vent = total_vent / valid_counts

print("Average ICU patients on ventilators per region:\n")
for r, avg in zip(unique_regions, avg_vent):
    print(r, ":", avg)
```

# <span style="color:Green">P9: Calculate the average number of patients per region who were in the ICU but not on ventilators ('icu_current_covid'-'icu_current_covid_vented'). </span>

```python
import numpy as np

region_key = "oh_region"
icu_key = "icu_current_covid"
vent_key = "icu_current_covid_vented"

# Extract regions
regions = np.array([row[region_key] for row in data], dtype=str)

# Clean ICU values
icu = np.array([
    float(row[icu_key]) if row[icu_key] not in ('.', '', 'NA') else np.nan
    for row in data
], dtype=float)

# Clean ventilator values
vent = np.array([
    float(row[vent_key]) if row[vent_key] not in ('.', '', 'NA') else np.nan
    for row in data
], dtype=float)

# ICU patients NOT on ventilators
icu_not_vented = icu - vent

# Group by region
unique_regions, inverse_idx = np.unique(regions, return_inverse=True)

# Sum values per region (ignore NaN)
total_not_vented = np.bincount(inverse_idx, weights=np.nan_to_num(icu_not_vented))

# Count valid values per region
valid_counts = np.bincount(inverse_idx, weights=~np.isnan(icu_not_vented))

# Average per region
avg_not_vented = total_not_vented / valid_counts

print("Average ICU patients NOT on ventilators per region:\n")
for r, avg in zip(unique_regions, avg_not_vented):
    print(r, ":", avg)
```

# <span style="color:Green">P10: What are the top 5 days with the highest number of ICU COVID ('icu_current_covid') cases? </span>

```python
#Codes of P10 here
import numpy as np

date_key = "date"
icu_key = "icu_current_covid"

# Extract date column
dates = np.array([row[date_key] for row in data], dtype=str)

# Clean ICU values
icu = np.array([
    float(row[icu_key]) if row[icu_key] not in ('.', '', 'NA') else np.nan
    for row in data
], dtype=float)

# Group by date
unique_dates, inverse_idx = np.unique(dates, return_inverse=True)

# Sum ICU per day (ignore NaN)
daily_totals = np.bincount(inverse_idx, weights=np.nan_to_num(icu))

# Sort descending
sorted_indices = np.argsort(daily_totals)[::-1]

# Top 5 days
top5 = sorted_indices[:5]

print("Top 5 days with highest ICU COVID cases:\n")

for idx in top5:
    print("Date:", unique_dates[idx], "| Total ICU:", daily_totals[idx])
```
